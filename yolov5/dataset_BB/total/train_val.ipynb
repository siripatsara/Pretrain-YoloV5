```python
import os
import random
import shutil

# กำหนด path
img_dir = r'C:/university/241-353/T.Fern/program/yolov5/dataset_BB/total/Images'
train_dir = r'C:/university/241-353/T.Fern/program/yolov5/dataset_BB/total/images/train'
val_dir = r'C:/university/241-353/T.Fern/program/yolov5/dataset_BB/total/images/val'

# สร้างโฟลเดอร์ถ้ายังไม่มี
os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)

# ดึงรายชื่อไฟล์ภาพ
img_files = [f for f in os.listdir(img_dir) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
random.shuffle(img_files)

# แบ่ง 80% train, 20% val
n_train = int(len(img_files) * 0.8)
train_files = img_files[:n_train]
val_files = img_files[n_train:]

# คัดลอกไฟล์
for f in train_files:
    shutil.copy(os.path.join(img_dir, f), os.path.join(train_dir, f))
for f in val_files:
    shutil.copy(os.path.join(img_dir, f), os.path.join(val_dir, f))

print(f'แบ่ง train: {len(train_files)} val: {len(val_files)} สำเร็จ!')

# แบ่ง Image สำหรับ train และ val
```

```python
import os

# Path ไปยังไฟล์และโฟลเดอร์
train_txt_path = r'C:/university/241-353/T.Fern/program/yolov5/dataset_BB/total/train.txt'
train_img_dir = r'C:/university/241-353/T.Fern/program/yolov5/dataset_BB/total/images/train'
val_img_dir = r'C:/university/241-353/T.Fern/program/yolov5/dataset_BB/total/images/val'

# อ่านรายชื่อไฟล์ภาพใน train และ val (ไม่รวม path)
train_imgs = set(os.listdir(train_img_dir))
val_imgs = set(os.listdir(val_img_dir))

# อ่านไฟล์ train.txt (แต่ละบรรทัดคือ path หรือชื่อไฟล์)
with open(train_txt_path, 'r', encoding='utf-8') as f:
    lines = [line.strip() for line in f if line.strip()]

# แยกไฟล์ที่อยู่ใน train และ val
train_lines = [line for line in lines if os.path.basename(line) in train_imgs]
val_lines = [line for line in lines if os.path.basename(line) in val_imgs]

# เขียนไฟล์ใหม่
with open('train_split.txt', 'w', encoding='utf-8') as f:
    for line in train_lines:
        f.write(line + '\n')

with open('val_split.txt', 'w', encoding='utf-8') as f:
    for line in val_lines:
        f.write(line + '\n')

print(f'สร้างไฟล์ train_split.txt ({len(train_lines)} บรรทัด) และ val_split.txt ({len(val_lines)} บรรทัด) สำเร็จ!')

# แบ่ง train และ val จาก train.txt แต่ต้องลบไฟล์นี้ออกด้วยนะ
```

```python
# ตรวจสอบว่า train_split.txt และ val_split.txt มีไฟล์ซ้ำกันหรือไม่
train_split_path = r'C:/university/241-353/T.Fern/program/yolov5/dataset_BB/total/train_split.txt'
val_split_path = r'C:/university/241-353/T.Fern/program/yolov5/dataset_BB/total/val_split.txt'

with open(train_split_path, 'r', encoding='utf-8') as f:
    train_lines = set([os.path.basename(line.strip()) for line in f if line.strip()])

with open(val_split_path, 'r', encoding='utf-8') as f:
    val_lines = set([os.path.basename(line.strip()) for line in f if line.strip()])

duplicates = train_lines.intersection(val_lines)

if duplicates:
    print(f'พบไฟล์ซ้ำ {len(duplicates)} ไฟล์ใน train_split.txt และ val_split.txt:')
    for f in duplicates:
        print(f)
else:
    print('ไม่พบไฟล์ซ้ำระหว่าง train_split.txt และ val_split.txt')
    

# ตรวจสอบว่า ข้อมูลใน train_split.txt และ val_split.txt มีไฟล์ซ้ำกันหรือไม่
```

```python
import os
import shutil

# Path
train_img_dir = r'C:/university/241-353/T.Fern/program/yolov5/dataset_BB/total/images/train'
val_img_dir = r'C:/university/241-353/T.Fern/program/yolov5/dataset_BB/total/images/val'
label_src_dir = r'C:/university/241-353/T.Fern/program/yolov5/dataset_BB/total/labels/train'
label_train_dst = r'C:/university/241-353/T.Fern/program/yolov5/dataset_BB/total/labels_split/train'
label_val_dst = r'C:/university/241-353/T.Fern/program/yolov5/dataset_BB/total/labels_split/val'

os.makedirs(label_train_dst, exist_ok=True)
os.makedirs(label_val_dst, exist_ok=True)

# รายชื่อไฟล์ภาพ (ไม่เอานามสกุล)
train_basenames = {os.path.splitext(f)[0] for f in os.listdir(train_img_dir)}
val_basenames = {os.path.splitext(f)[0] for f in os.listdir(val_img_dir)}

# คัดลอก label ที่ตรงกับ train
for name in train_basenames:
    label_file = os.path.join(label_src_dir, name + '.txt')
    if os.path.exists(label_file):
        shutil.copy(label_file, os.path.join(label_train_dst, name + '.txt'))

# คัดลอก label ที่ตรงกับ val
for name in val_basenames:
    label_file = os.path.join(label_src_dir, name + '.txt')
    if os.path.exists(label_file):
        shutil.copy(label_file, os.path.join(label_val_dst, name + '.txt'))

print('คัดลอก label สำหรับ train และ val สำเร็จ!')

# ทำการแบ่ง trai และ val ให้เหมือนใน image สำหรับ label
```

```python
# ตรวจสอบ path ในไฟล์ train.txt และ val.txt ว่าตรงกับไฟล์ภาพจริงหรือไม่
def check_image_paths(txt_path, img_root):
    import os
    missing = []
    with open(txt_path, 'r', encoding='utf-8') as f:
        for line in f:
            path = line.strip()
            if not path: continue
            # ถ้า path เป็น relative path ให้ต่อกับ img_root
            img_path = os.path.join(img_root, path) if not os.path.isabs(path) else path
            if not os.path.exists(img_path):
                missing.append(img_path)
    return missing

# Path ไปยังไฟล์และโฟลเดอร์ภาพ
train_txt = r'C:/university/241-353/T.Fern/program/yolov5/dataset_BB/total/train.txt'
val_txt = r'C:/university/241-353/T.Fern/program/yolov5/dataset_BB/total/val.txt'
img_root = r'C:/university/241-353/T.Fern/program/yolov5/dataset_BB/total'  # root หลักที่เก็บ Images

missing_train = check_image_paths(train_txt, img_root)
missing_val = check_image_paths(val_txt, img_root)

if missing_train:
    print(f'พบ path ใน train.txt ที่ไม่เจอไฟล์จริง {len(missing_train)} ไฟล์:')
    for p in missing_train:
        print(p)
else:
    print('train.txt: ทุก path ชี้ไปยังไฟล์ภาพจริง')

if missing_val:
    print(f'พบ path ใน val.txt ที่ไม่เจอไฟล์จริง {len(missing_val)} ไฟล์:')
    for p in missing_val:
        print(p)
else:
    print('val.txt: ทุก path ชี้ไปยังไฟล์ภาพจริง')
```
