```python
import os

# Path ‡πÑ‡∏õ‡∏¢‡∏±‡∏á‡πÑ‡∏ü‡∏•‡πå‡πÅ‡∏•‡∏∞‡πÇ‡∏ü‡∏•‡πÄ‡∏î‡∏≠‡∏£‡πå
train_txt_path = r'C:/university/241-353/T.Fern/program/yolov5/dataset_allBB/train.txt'
train_img_dir = r'C:/university/241-353/T.Fern/program/yolov5/dataset_allBB/images/train'
val_img_dir = r'C:/university/241-353/T.Fern/program/yolov5/dataset_allBB/images/val'

# ‡∏≠‡πà‡∏≤‡∏ô‡∏£‡∏≤‡∏¢‡∏ä‡∏∑‡πà‡∏≠‡πÑ‡∏ü‡∏•‡πå‡∏†‡∏≤‡∏û‡πÉ‡∏ô train ‡πÅ‡∏•‡∏∞ val (‡πÑ‡∏°‡πà‡∏£‡∏ß‡∏° path)
train_imgs = set(os.listdir(train_img_dir))
val_imgs = set(os.listdir(val_img_dir))

# ‡∏≠‡πà‡∏≤‡∏ô‡πÑ‡∏ü‡∏•‡πå train.txt (‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î‡∏Ñ‡∏∑‡∏≠ path ‡∏´‡∏£‡∏∑‡∏≠‡∏ä‡∏∑‡πà‡∏≠‡πÑ‡∏ü‡∏•‡πå)
with open(train_txt_path, 'r', encoding='utf-8') as f:
    lines = [line.strip() for line in f if line.strip()]

# ‡πÅ‡∏¢‡∏Å‡πÑ‡∏ü‡∏•‡πå‡∏ó‡∏µ‡πà‡∏≠‡∏¢‡∏π‡πà‡πÉ‡∏ô train ‡πÅ‡∏•‡∏∞ val
train_lines = [line for line in lines if os.path.basename(line) in train_imgs]
val_lines = [line for line in lines if os.path.basename(line) in val_imgs]

# ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡πÑ‡∏ü‡∏•‡πå‡πÉ‡∏´‡∏°‡πà
with open('train_split.txt', 'w', encoding='utf-8') as f:
    for line in train_lines:
        f.write(line + '\n')

with open('val_split.txt', 'w', encoding='utf-8') as f:
    for line in val_lines:
        f.write(line + '\n')

print(f'‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÑ‡∏ü‡∏•‡πå train_split.txt ({len(train_lines)} ‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î) ‡πÅ‡∏•‡∏∞ val_split.txt ({len(val_lines)} ‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î) ‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à!')

# ‡πÅ‡∏ö‡πà‡∏á train ‡πÅ‡∏•‡∏∞ val ‡∏à‡∏≤‡∏Å train.txt ‡πÅ‡∏ï‡πà‡∏ï‡πâ‡∏≠‡∏á‡∏•‡∏ö‡πÑ‡∏ü‡∏•‡πå‡∏ô‡∏µ‡πâ‡∏≠‡∏≠‡∏Å‡∏î‡πâ‡∏ß‡∏¢‡∏ô‡∏∞
```

```python
# ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ß‡πà‡∏≤ train_split.txt ‡πÅ‡∏•‡∏∞ val_split.txt ‡∏°‡∏µ‡πÑ‡∏ü‡∏•‡πå‡∏ã‡πâ‡∏≥‡∏Å‡∏±‡∏ô‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà
train_split_path = r'C:/university/241-353/T.Fern/program/yolov5/dataset_allBB/train_split.txt'
val_split_path = r'C:/university/241-353/T.Fern/program/yolov5/dataset_allBB/val_split.txt'

with open(train_split_path, 'r', encoding='utf-8') as f:
    train_lines = set([os.path.basename(line.strip()) for line in f if line.strip()])

with open(val_split_path, 'r', encoding='utf-8') as f:
    val_lines = set([os.path.basename(line.strip()) for line in f if line.strip()])

duplicates = train_lines.intersection(val_lines)

if duplicates:
    print(f'‡∏û‡∏ö‡πÑ‡∏ü‡∏•‡πå‡∏ã‡πâ‡∏≥ {len(duplicates)} ‡πÑ‡∏ü‡∏•‡πå‡πÉ‡∏ô train_split.txt ‡πÅ‡∏•‡∏∞ val_split.txt:')
    for f in duplicates:
        print(f)
else:
    print('‡πÑ‡∏°‡πà‡∏û‡∏ö‡πÑ‡∏ü‡∏•‡πå‡∏ã‡πâ‡∏≥‡∏£‡∏∞‡∏´‡∏ß‡πà‡∏≤‡∏á train_split.txt ‡πÅ‡∏•‡∏∞ val_split.txt')
    

# ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ß‡πà‡∏≤ ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡πÉ‡∏ô train_split.txt ‡πÅ‡∏•‡∏∞ val_split.txt ‡∏°‡∏µ‡πÑ‡∏ü‡∏•‡πå‡∏ã‡πâ‡∏≥‡∏Å‡∏±‡∏ô‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà
```

```python
import os
import shutil

# Path
train_img_dir = r'C:/university/241-353/T.Fern/program/yolov5/dataset_allBB/images/train'
val_img_dir = r'C:/university/241-353/T.Fern/program/yolov5/dataset_allBB/images/val'
label_src_dir = r'C:/university/241-353/T.Fern/program/yolov5/dataset_allBB/labels/train'
label_train_dst = r'C:/university/241-353/T.Fern/program/yolov5/dataset_allBB/labels_split/train'
label_val_dst = r'C:/university/241-353/T.Fern/program/yolov5/dataset_allBB/labels_split/val'

os.makedirs(label_train_dst, exist_ok=True)
os.makedirs(label_val_dst, exist_ok=True)

# ‡∏£‡∏≤‡∏¢‡∏ä‡∏∑‡πà‡∏≠‡πÑ‡∏ü‡∏•‡πå‡∏†‡∏≤‡∏û (‡πÑ‡∏°‡πà‡πÄ‡∏≠‡∏≤‡∏ô‡∏≤‡∏°‡∏™‡∏Å‡∏∏‡∏•)
train_basenames = {os.path.splitext(f)[0] for f in os.listdir(train_img_dir)}
val_basenames = {os.path.splitext(f)[0] for f in os.listdir(val_img_dir)}

# ‡∏Ñ‡∏±‡∏î‡∏•‡∏≠‡∏Å label ‡∏ó‡∏µ‡πà‡∏ï‡∏£‡∏á‡∏Å‡∏±‡∏ö train
for name in train_basenames:
    label_file = os.path.join(label_src_dir, name + '.txt')
    if os.path.exists(label_file):
        shutil.copy(label_file, os.path.join(label_train_dst, name + '.txt'))

# ‡∏Ñ‡∏±‡∏î‡∏•‡∏≠‡∏Å label ‡∏ó‡∏µ‡πà‡∏ï‡∏£‡∏á‡∏Å‡∏±‡∏ö val
for name in val_basenames:
    label_file = os.path.join(label_src_dir, name + '.txt')
    if os.path.exists(label_file):
        shutil.copy(label_file, os.path.join(label_val_dst, name + '.txt'))

print('‡∏Ñ‡∏±‡∏î‡∏•‡∏≠‡∏Å label ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö train ‡πÅ‡∏•‡∏∞ val ‡∏™‡∏≥‡πÄ‡∏£‡πá‡∏à!')

# ‡∏ó‡∏≥‡∏Å‡∏≤‡∏£‡πÅ‡∏ö‡πà‡∏á train ‡πÅ‡∏•‡∏∞ val ‡πÉ‡∏´‡πâ‡πÄ‡∏´‡∏°‡∏∑‡∏≠‡∏ô‡πÉ‡∏ô image ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö label
```

```python
import os

label_train_dir = r'C:/university/241-353/T.Fern/program/yolov5/dataset_allBB/labels_split/train'
label_val_dir = r'C:/university/241-353/T.Fern/program/yolov5/dataset_allBB/labels_split/val'

train_labels = set(os.listdir(label_train_dir))
val_labels = set(os.listdir(label_val_dir))

duplicates = train_labels.intersection(val_labels)

if duplicates:
    print(f'‡∏û‡∏ö‡πÑ‡∏ü‡∏•‡πå label ‡∏ã‡πâ‡∏≥ {len(duplicates)} ‡πÑ‡∏ü‡∏•‡πå:')
    for f in duplicates:
        print(f)
else:
    print('‡πÑ‡∏°‡πà‡∏û‡∏ö‡πÑ‡∏ü‡∏•‡πå label ‡∏ã‡πâ‡∏≥‡∏£‡∏∞‡∏´‡∏ß‡πà‡∏≤‡∏á train ‡πÅ‡∏•‡∏∞ val')
    
# ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ß‡πà‡∏≤‡πÉ‡∏ô label ‡∏°‡∏µ train ‡πÅ‡∏•‡∏∞ val ‡∏ã‡πâ‡∏≥‡∏Å‡∏±‡∏ô‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà
```

```python
import os

def check_paths_in_file(txt_file_path, img_root):
    """‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ß‡πà‡∏≤ path ‡πÉ‡∏ô‡πÑ‡∏ü‡∏•‡πå txt ‡∏ä‡∏µ‡πâ‡πÑ‡∏õ‡∏¢‡∏±‡∏á‡πÑ‡∏ü‡∏•‡πå‡∏†‡∏≤‡∏û‡∏à‡∏£‡∏¥‡∏á‡πÜ ‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà"""
    print(f"\n=== ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡πÑ‡∏ü‡∏•‡πå: {os.path.basename(txt_file_path)} ===")
    
    if not os.path.exists(txt_file_path):
        print(f"‚ùå ‡πÑ‡∏°‡πà‡∏û‡∏ö‡πÑ‡∏ü‡∏•‡πå {txt_file_path}")
        return
    
    with open(txt_file_path, 'r', encoding='utf-8') as f:
        lines = [line.strip() for line in f if line.strip()]
    
    print(f"üìù ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î: {len(lines)}")
    
    # ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö path ‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î
    existing_files = []
    missing_files = []
    
    for i, line in enumerate(lines[:10]):  # ‡πÅ‡∏™‡∏î‡∏á‡πÅ‡∏Ñ‡πà 10 ‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î‡πÅ‡∏£‡∏Å
        # ‡∏•‡∏≠‡∏á‡∏´‡∏•‡∏≤‡∏¢‡∏ß‡∏¥‡∏ò‡∏µ‡πÉ‡∏ô‡∏Å‡∏≤‡∏£‡∏™‡∏£‡πâ‡∏≤‡∏á full path
        possible_paths = [
            os.path.join(img_root, line),  # path ‡∏ï‡∏£‡∏á‡πÜ
            os.path.join(img_root, line.replace('/', os.sep)),  # ‡πÅ‡∏õ‡∏•‡∏á / ‡πÄ‡∏õ‡πá‡∏ô \
            line if os.path.isabs(line) else os.path.join(img_root, line)  # ‡∏ñ‡πâ‡∏≤‡πÄ‡∏õ‡πá‡∏ô absolute path
        ]
        
        found = False
        for full_path in possible_paths:
            if os.path.exists(full_path):
                existing_files.append((line, full_path))
                print(f"‚úÖ [{i+1:2d}] {line} ‚Üí {full_path}")
                found = True
                break
        
        if not found:
            missing_files.append(line)
            print(f"‚ùå [{i+1:2d}] {line} ‚Üí ‡πÑ‡∏°‡πà‡∏û‡∏ö‡πÑ‡∏ü‡∏•‡πå")
    
    if len(lines) > 10:
        print(f"... ‡πÅ‡∏•‡∏∞‡∏≠‡∏µ‡∏Å {len(lines) - 10} ‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î")
    
    # ‡∏™‡∏£‡∏∏‡∏õ‡∏ú‡∏•
    total_existing = 0
    total_missing = 0
    
    for line in lines:
        possible_paths = [
            os.path.join(img_root, line),
            os.path.join(img_root, line.replace('/', os.sep)),
            line if os.path.isabs(line) else os.path.join(img_root, line)
        ]
        
        found = any(os.path.exists(p) for p in possible_paths)
        if found:
            total_existing += 1
        else:
            total_missing += 1
    
    print(f"\nüìä ‡∏™‡∏£‡∏∏‡∏õ:")
    print(f"‚úÖ ‡πÑ‡∏ü‡∏•‡πå‡∏ó‡∏µ‡πà‡∏û‡∏ö: {total_existing}")
    print(f"‚ùå ‡πÑ‡∏ü‡∏•‡πå‡∏ó‡∏µ‡πà‡πÑ‡∏°‡πà‡∏û‡∏ö: {total_missing}")
    print(f"üìà ‡πÄ‡∏õ‡∏≠‡∏£‡πå‡πÄ‡∏ã‡πá‡∏ô‡∏ï‡πå‡∏ó‡∏µ‡πà‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á: {(total_existing/len(lines)*100):.1f}%")
    
    return total_existing, total_missing

# ‡∏Å‡∏≥‡∏´‡∏ô‡∏î path
img_root = r'C:/university/241-353/T.Fern/program/yolov5/dataset_allBB'
train_txt_path = os.path.join(img_root, 'train.txt')
val_txt_path = os.path.join(img_root, 'val.txt')

# ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡πÑ‡∏ü‡∏•‡πå train.txt
train_ok, train_missing = check_paths_in_file(train_txt_path, img_root)

# ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡πÑ‡∏ü‡∏•‡πå val.txt
val_ok, val_missing = check_paths_in_file(val_txt_path, img_root)

print(f"\nüéØ ‡∏™‡∏£‡∏∏‡∏õ‡∏£‡∏ß‡∏°:")
print(f"Train: {train_ok} ‡∏ñ‡∏π‡∏Å, {train_missing} ‡∏ú‡∏¥‡∏î")
print(f"Val: {val_ok} ‡∏ñ‡∏π‡∏Å, {val_missing} ‡∏ú‡∏¥‡∏î")

# ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ß‡πà‡∏≤ train.txt ‡πÅ‡∏•‡∏∞ val.txt ‡∏ä‡∏µ‡πâ‡πÑ‡∏õ‡∏¢‡∏±‡∏á path ‡∏†‡∏≤‡∏û‡∏ñ‡∏π‡∏Å‡∏ï‡πâ‡∏≠‡∏á‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà
```

```python
import os

def fix_paths_in_txt_file(file_path):
    """‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç path ‡πÉ‡∏ô‡πÑ‡∏ü‡∏•‡πå txt ‡πÇ‡∏î‡∏¢‡πÄ‡∏≠‡∏≤ dataset_allBB/ ‡∏≠‡∏≠‡∏Å‡∏à‡∏≤‡∏Å‡∏´‡∏ô‡πâ‡∏≤"""
    print(f"üîß ‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç‡πÑ‡∏ü‡∏•‡πå: {os.path.basename(file_path)}")
    
    if not os.path.exists(file_path):
        print(f"‚ùå ‡πÑ‡∏°‡πà‡∏û‡∏ö‡πÑ‡∏ü‡∏•‡πå {file_path}")
        return
    
    # ‡∏≠‡πà‡∏≤‡∏ô‡πÑ‡∏ü‡∏•‡πå‡πÄ‡∏î‡∏¥‡∏°
    with open(file_path, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    
    print(f"üìù ‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î‡πÄ‡∏î‡∏¥‡∏°: {len(lines)}")
    
    # ‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç path
    fixed_lines = []
    for line in lines:
        line = line.strip()
        if line.startswith('dataset_allBB/'):
            # ‡πÄ‡∏≠‡∏≤ dataset_allBB/ ‡∏≠‡∏≠‡∏Å‡∏à‡∏≤‡∏Å‡∏´‡∏ô‡πâ‡∏≤
            new_line = line.replace('dataset_allBB/', '', 1)  # ‡πÅ‡∏ó‡∏ô‡∏ó‡∏µ‡πà‡πÅ‡∏Ñ‡πà‡∏Ñ‡∏£‡∏±‡πâ‡∏á‡πÅ‡∏£‡∏Å
            fixed_lines.append(new_line + '\n')
            print(f"‚úèÔ∏è  {line} ‚Üí {new_line}")
        else:
            fixed_lines.append(line + '\n')
    
    # ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡πÑ‡∏ü‡∏•‡πå‡πÉ‡∏´‡∏°‡πà
    with open(file_path, 'w', encoding='utf-8') as f:
        f.writelines(fixed_lines)
    
    print(f"‚úÖ ‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç‡πÄ‡∏™‡∏£‡πá‡∏à‡πÅ‡∏•‡πâ‡∏ß!")

# ‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç‡πÑ‡∏ü‡∏•‡πå train.txt ‡πÅ‡∏•‡∏∞ val.txt
img_root = r'C:/university/241-353/T.Fern/program/yolov5/dataset_allBB'
train_txt_path = os.path.join(img_root, 'train.txt')
val_txt_path = os.path.join(img_root, 'val.txt')

fix_paths_in_txt_file(train_txt_path)
fix_paths_in_txt_file(val_txt_path)

print("\nüéâ ‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç path ‡πÉ‡∏ô‡πÑ‡∏ü‡∏•‡πå train.txt ‡πÅ‡∏•‡∏∞ val.txt ‡πÄ‡∏™‡∏£‡πá‡∏à‡πÅ‡∏•‡πâ‡∏ß!")

# ‡πÅ‡∏Å‡πâ‡πÑ‡∏Ç path ‡πÉ‡∏ô‡πÑ‡∏ü‡∏•‡πå train.txt ‡πÅ‡∏•‡∏∞ val.txt
```
