```python
import os
import random
import shutil

# กำหนด path
img_dir = r'C:/university/241-353/T.Fern/program/yolov5/dataset_allBB/Images'
train_dir = r'C:/university/241-353/T.Fern/program/yolov5/dataset_allBB/images/train'
val_dir = r'C:/university/241-353/T.Fern/program/yolov5/dataset_allBB/images/val'

# สร้างโฟลเดอร์ถ้ายังไม่มี
os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)

# ดึงรายชื่อไฟล์ภาพ
img_files = [f for f in os.listdir(img_dir) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
random.shuffle(img_files)

# แบ่ง 80% train, 20% val
n_train = int(len(img_files) * 0.8)
train_files = img_files[:n_train]
val_files = img_files[n_train:]

# คัดลอกไฟล์
for f in train_files:
    shutil.copy(os.path.join(img_dir, f), os.path.join(train_dir, f))
for f in val_files:
    shutil.copy(os.path.join(img_dir, f), os.path.join(val_dir, f))

print(f'แบ่ง train: {len(train_files)} val: {len(val_files)} สำเร็จ!')

# แบ่ง Image สำหรับ train และ val
```

```python
import os

# Path ไปยังไฟล์และโฟลเดอร์
train_txt_path = r'C:/university/241-353/T.Fern/program/yolov5/dataset_allBB/train.txt'
train_img_dir = r'C:/university/241-353/T.Fern/program/yolov5/dataset_allBB/images/train'
val_img_dir = r'C:/university/241-353/T.Fern/program/yolov5/dataset_allBB/images/val'

# อ่านรายชื่อไฟล์ภาพใน train และ val (ไม่รวม path)
train_imgs = set(os.listdir(train_img_dir))
val_imgs = set(os.listdir(val_img_dir))

# อ่านไฟล์ train.txt (แต่ละบรรทัดคือ path หรือชื่อไฟล์)
with open(train_txt_path, 'r', encoding='utf-8') as f:
    lines = [line.strip() for line in f if line.strip()]

# แยกไฟล์ที่อยู่ใน train และ val
train_lines = [line for line in lines if os.path.basename(line) in train_imgs]
val_lines = [line for line in lines if os.path.basename(line) in val_imgs]

# เขียนไฟล์ใหม่
with open('train_split.txt', 'w', encoding='utf-8') as f:
    for line in train_lines:
        f.write(line + '\n')

with open('val_split.txt', 'w', encoding='utf-8') as f:
    for line in val_lines:
        f.write(line + '\n')

print(f'สร้างไฟล์ train_split.txt ({len(train_lines)} บรรทัด) และ val_split.txt ({len(val_lines)} บรรทัด) สำเร็จ!')

# แบ่ง train และ val จาก train.txt แต่ต้องลบไฟล์นี้ออกด้วยนะ
```

```python
import os
import shutil

# Path
train_img_dir = r'C:/university/241-353/T.Fern/program/yolov5/dataset_allBB/images/train'
val_img_dir = r'C:/university/241-353/T.Fern/program/yolov5/dataset_allBB/images/val'
label_src_dir = r'C:/university/241-353/T.Fern/program/yolov5/dataset_allBB/labels/train'
label_train_dst = r'C:/university/241-353/T.Fern/program/yolov5/dataset_allBB/labels_split/train'
label_val_dst = r'C:/university/241-353/T.Fern/program/yolov5/dataset_allBB/labels_split/val'

os.makedirs(label_train_dst, exist_ok=True)
os.makedirs(label_val_dst, exist_ok=True)

# รายชื่อไฟล์ภาพ (ไม่เอานามสกุล)
train_basenames = {os.path.splitext(f)[0] for f in os.listdir(train_img_dir)}
val_basenames = {os.path.splitext(f)[0] for f in os.listdir(val_img_dir)}

# คัดลอก label ที่ตรงกับ train
for name in train_basenames:
    label_file = os.path.join(label_src_dir, name + '.txt')
    if os.path.exists(label_file):
        shutil.copy(label_file, os.path.join(label_train_dst, name + '.txt'))

# คัดลอก label ที่ตรงกับ val
for name in val_basenames:
    label_file = os.path.join(label_src_dir, name + '.txt')
    if os.path.exists(label_file):
        shutil.copy(label_file, os.path.join(label_val_dst, name + '.txt'))

print('คัดลอก label สำหรับ train และ val สำเร็จ!')

# ทำการแบ่ง train และ val ให้เหมือนใน image สำหรับ label
```

```python
import os

def check_paths_in_file(txt_file_path, img_root):
    """ตรวจสอบว่า path ในไฟล์ txt ชี้ไปยังไฟล์ภาพจริงๆ หรือไม่"""
    print(f"\n=== ตรวจสอบไฟล์: {os.path.basename(txt_file_path)} ===")
    
    if not os.path.exists(txt_file_path):
        print(f"❌ ไม่พบไฟล์ {txt_file_path}")
        return
    
    with open(txt_file_path, 'r', encoding='utf-8') as f:
        lines = [line.strip() for line in f if line.strip()]
    
    print(f"📝 จำนวนบรรทัดทั้งหมด: {len(lines)}")
    
    # ตรวจสอบ path แต่ละบรรทัด
    existing_files = []
    missing_files = []
    
    for i, line in enumerate(lines[:10]):  # แสดงแค่ 10 บรรทัดแรก
        # ลองหลายวิธีในการสร้าง full path
        possible_paths = [
            os.path.join(img_root, line),  # path ตรงๆ
            os.path.join(img_root, line.replace('/', os.sep)),  # แปลง / เป็น \
            line if os.path.isabs(line) else os.path.join(img_root, line)  # ถ้าเป็น absolute path
        ]
        
        found = False
        for full_path in possible_paths:
            if os.path.exists(full_path):
                existing_files.append((line, full_path))
                print(f"✅ [{i+1:2d}] {line} → {full_path}")
                found = True
                break
        
        if not found:
            missing_files.append(line)
            print(f"❌ [{i+1:2d}] {line} → ไม่พบไฟล์")
    
    if len(lines) > 10:
        print(f"... และอีก {len(lines) - 10} บรรทัด")
    
    # สรุปผล
    total_existing = 0
    total_missing = 0
    
    for line in lines:
        possible_paths = [
            os.path.join(img_root, line),
            os.path.join(img_root, line.replace('/', os.sep)),
            line if os.path.isabs(line) else os.path.join(img_root, line)
        ]
        
        found = any(os.path.exists(p) for p in possible_paths)
        if found:
            total_existing += 1
        else:
            total_missing += 1
    
    print(f"\n📊 สรุป:")
    print(f"✅ ไฟล์ที่พบ: {total_existing}")
    print(f"❌ ไฟล์ที่ไม่พบ: {total_missing}")
    print(f"📈 เปอร์เซ็นต์ที่ถูกต้อง: {(total_existing/len(lines)*100):.1f}%")
    
    return total_existing, total_missing

# กำหนด path
img_root = r'C:/university/241-353/T.Fern/program/yolov5/dataset_allBB'
train_txt_path = os.path.join(img_root, 'train.txt')
val_txt_path = os.path.join(img_root, 'val.txt')

# ตรวจสอบไฟล์ train.txt
train_ok, train_missing = check_paths_in_file(train_txt_path, img_root)

# ตรวจสอบไฟล์ val.txt
val_ok, val_missing = check_paths_in_file(val_txt_path, img_root)

print(f"\n🎯 สรุปรวม:")
print(f"Train: {train_ok} ถูก, {train_missing} ผิด")
print(f"Val: {val_ok} ถูก, {val_missing} ผิด")

# ตรวจสอบว่า train.txt และ val.txt ชี้ไปยัง path ภาพถูกต้องหรือไม่
```

```python
import os

def fix_paths_in_txt_file(file_path):
    """แก้ไข path ในไฟล์ txt โดยเอา dataset_allBB/ ออกจากหน้า"""
    print(f"🔧 แก้ไขไฟล์: {os.path.basename(file_path)}")
    
    if not os.path.exists(file_path):
        print(f"❌ ไม่พบไฟล์ {file_path}")
        return
    
    # อ่านไฟล์เดิม
    with open(file_path, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    
    print(f"📝 จำนวนบรรทัดเดิม: {len(lines)}")
    
    # แก้ไข path
    fixed_lines = []
    for line in lines:
        line = line.strip()
        if line.startswith('dataset_allBB/'):
            # เอา dataset_allBB/ ออกจากหน้า
            new_line = line.replace('dataset_allBB/', '', 1)  # แทนที่แค่ครั้งแรก
            fixed_lines.append(new_line + '\n')
            print(f"✏️  {line} → {new_line}")
        else:
            fixed_lines.append(line + '\n')
    
    # เขียนไฟล์ใหม่
    with open(file_path, 'w', encoding='utf-8') as f:
        f.writelines(fixed_lines)
    
    print(f"✅ แก้ไขเสร็จแล้ว!")

# แก้ไขไฟล์ train.txt และ val.txt
img_root = r'C:/university/241-353/T.Fern/program/yolov5/dataset_allBB'
train_txt_path = os.path.join(img_root, 'train.txt')
val_txt_path = os.path.join(img_root, 'val.txt')

fix_paths_in_txt_file(train_txt_path)
fix_paths_in_txt_file(val_txt_path)

print("\n🎉 แก้ไข path ในไฟล์ train.txt และ val.txt เสร็จแล้ว!")

# แก้ไข path ในไฟล์ train.txt และ val.txt
```
